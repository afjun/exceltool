## 身份证号校验

```python
def check_id_card(num: str):
    num.replace(' ' , '')
    data_num = [7,9,10,5,8,4,2,1,6,3,7,9,10,5,8,4,2]
    sum = 0
    for i in range(17):
        try:
            sum += int(num[i])*data_num[i]
        except ValueError:
            return False
    result=sum%11
    check_dict={ 0:1,1:0,2:"X",3:9,4:8,5:7,6:6,7:5,8:4,9:3,10:2}
    if result in check_dict.keys():
        if str(check_dict.get(result)) == str(num[-1]):
            print(num)
            return True
        else:
            return False
    return False
```

## 手机号校验

```python
def check_phone(num:str):
    if num[-4:] == "0000":
        return False
    if len(num)==11:
        return True
    return False
        
```

```python
# 按行获取值
rows = []
for i in sheet.iter_rows(min_row=2, max_row=howmany_rows, min_col=1, max_col=14):
    row = []
    for cell in i:
        row.append(cell.value)
    temp_rows = [] #每一行分完放这里
    try:
        row[6] = row[6].replace(" ", "")#去掉空格
    except AttributeError:
        break
    timeList = row[6].split("到")
    print(timeList)
    if(len(timeList)>1):#此时需要进行时间分割
        datetime_str = '09/19/18 13:55:26'
        start_date_time = datetime.strptime(timeList[0], '%Y-%m-%d')
        end_date_time = datetime.strptime(timeList[1], '%Y-%m-%d')
#         print("start_date_time:")
#         print(start_date_time)
#         print("end_date_time:")
#         print(end_date_time)
        for day in range(start_date_time.day,end_date_time.day+1):
#             print(day)
            if len(str(day)) == 1:
                this_date_time = start_date_time.strftime('%Y-%m-0')
            else:
                this_date_time = start_date_time.strftime('%Y-%m-')


            this_date_time += str(day)
#             print("this_date_time:" + str(this_date_time))

            row[6] = this_date_time
#             print("row:" + str(row))
            temp_rows.append(copy.deepcopy(row))
            
#             print("this_date_time:" + str(this_date_time))
#             temp_rows[len(temp_rows)-1][6] = str(this_date_time)
    else:
        temp_rows.append(copy.deepcopy(row))
    print("temp_rows")
    print(temp_rows)
    
    
    
#     print(temp_rows[-1][6])
#     rows.append(row) #最后append到最终的rows里面
# rows

    print("----------------------------     楼宇分割    --------------------------------------")
    #楼宇分割
    temp_build_rows = []
    for row in temp_rows:
        row = copy.deepcopy(row)#新建一个row地址
#         small_build_list = row[9].split('[,|;*]+、')
        try:
        
        
            small_build_list = re.split('[,|;*、，]+',row[9])
        except TypeError:
            res_sheet.append(row)
            continue
        print(small_build_list)
        for one_build in small_build_list:
#             print(one_build)
#这里需要后序添加模糊搜索
            one_build = convey_build(one_build)
            row[9] = one_build
            print("one_build:" + str(one_build))
            if one_build == "":
                continue
            row[10] = buildDict[one_build]
            row[8] = attrDict[row[10]]
#             print(row[8])
#             print(row[9])
#             print(row[10])
#             print(row)
            temp_build_rows.append(copy.deepcopy(row))
    print(temp_build_rows)
    for one_row in temp_build_rows:
        res_sheet.append(one_row)
res_workbook.save("result.xlsx") 
```
